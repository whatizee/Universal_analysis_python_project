```python
#inserting the population data into total_population table in popu
%run data_processing.ipynb
import sqlite3
Population_df = data_process_population()
df_peace_index = web_scrapper_peace_index()
df_insurance = web_scrapper_insurance_companies()

connetion = sqlite3.connect("Universal_analysis.db")
Population_df.to_sql('total_population', connetion,if_exists='replace')
connetion.commit()
connetion.close()

connetion = sqlite3.connect("Universal_analysis.db")
df_peace_index.to_sql('peace_index', connetion,if_exists='replace')
connetion.commit()
connetion.close()

connetion = sqlite3.connect("Universal_analysis.db")
df_insurance.to_sql('df_insurance', connetion,if_exists='replace')
connetion.commit()
connetion.close()


def fetch_population():
    connection = sqlite3.connect('Universal_analysis.db')
    cursor = connection.cursor()
    # Retrieve the total population from the 'total_population' table
    cursor.execute("SELECT * FROM total_population")
    total_population = cursor.fetchall()
    connection.close()
    return total_population

    
def fetch_peace_index():
    connection = sqlite3.connect('Universal_analysis.db')
    cursor = connection.cursor()
    # Retrieve the Peace Index from the 'peace_index' table
    cursor.execute("SELECT * FROM peace_index")
    peace_index = cursor.fetchall()
    connection.close()
    return peace_index
    
def fetch_insurance():
    connection = sqlite3.connect('Universal_analysis.db')
    cursor = connection.cursor()
    # Retrieve the Insurance company from the 'df_insurance' table
    cursor.execute("SELECT * FROM df_insurance")
    df_insurance = cursor.fetchall()
    connection.close()
    
    return df_insurance
```
