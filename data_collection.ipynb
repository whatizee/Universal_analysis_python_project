```python
def web_scrapper_population():
    print("Executing scraping population data from wikipedia........")
    import datetime    
    # Get the start time
    start_time = datetime.datetime.now()
    print("Start Time:", start_time)
    from bs4 import BeautifulSoup
    import pandas as pd
    import requests
    import sqlite3
    
    url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
    r = requests.get(url)

    if r.status_code == 200:
        soup = BeautifulSoup(r.text, "lxml")

        # Try to find the table by searching for a specific caption
        table = soup.find("table", {"class": "wikitable"})
        header = table.find_all("th")

        column_names = [th.get_text(strip=True).replace('–', '').replace('\xa0', ' ').strip() for th in header if th.get_text(strip=True) != '' and not th.get_text(strip=True).isnumeric()]
        Headings = [item for item in column_names if item != '']
        #print(Headings)
        Population_df = pd.DataFrame(columns=Headings)
        #print(Population_df)
        observations = table.find_all("tr")
        for i in observations[1:]:
            data = i.find_all("td")
            row_data = [column.get_text(strip=True) for column in data]
            row_data.pop()
            data.append(row_data)
            if len(row_data) == len(Headings):
                new_row = dict(zip(Headings, row_data))
                Population_df = pd.concat([Population_df, pd.DataFrame([new_row])], ignore_index=True)

        print("web scraping completed successfully......")
        print("Data Collected is given below")
        #display(Population_df)
        #print(Population_df.columns)
        
        

    else:
        print("Failed to retrieve the webpage. Status code:", r.status_code)


    return Population_df

```

```python
def web_scrapper_peace_index():
    print("Executing scraping population data from wikipedia........")
    import datetime    
    # Get the start time
    start_time = datetime.datetime.now()
    #print("Start Time:", start_time)
    from bs4 import BeautifulSoup
    import pandas as pd
    import requests
    import sqlite3
    
    url = "https://worldpopulationreview.com/country-rankings/safest-countries-in-the-world"
    r = requests.get(url)

    if r.status_code == 200:
        soup = BeautifulSoup(r.text, "lxml")
        table = soup.find("table", {"class": "tp-table-body is-narrow w-full min-w-full table-auto border-separate border-spacing-0 border bg-white"})
        #print(table)
        ol = table.find_all("tr")
        #print(ol[1].find("a").text)
        #print(soup)
        countries = []
        peace_indexes_2023 = []
        peace_indexes_2022 = []
        
        for li in ol[1:]:
            
            country_name = li.find("a").text
            #print(country_name)
            #print(li)
            #peace_index = float(li.text.split("—")[1].strip())            
            peace_index = li.find_all("td")
            
            #print(peace_index[1].text)
            
            countries.append(country_name)
            peace_indexes_2023.append(peace_index[1].text)
            peace_indexes_2022.append(peace_index[2].text)
            
            
        # Create a DataFrame
        df_peace_index = pd.DataFrame({"Country": countries, "Peace Index 2023": peace_indexes_2023, "Peace Index 2022":peace_indexes_2022})
        
        #print(df_peace_index)
        #print(countries)
    else:
        print("Failed to retrieve the webpage. Status code:", r.status_code)


    return df_peace_index
```

```python
def web_scrapper_insurance_companies():
    print("Executing scraping List of Canadian Insurance Companies from wikipedia........")
    import datetime    
    # Get the start time
    start_time = datetime.datetime.now()
    #print("Start Time:", start_time)
    from bs4 import BeautifulSoup
    import pandas as pd
    import requests
    import sqlite3
    url = "https://en.wikipedia.org/wiki/List_of_Canadian_insurance_companies"
    r = requests.get(url)
 
    if r.status_code == 200:
        soup = BeautifulSoup(r.text, "lxml")
       # print(soup)
 
        #Try to find the table by searching for a specific caption
        table = soup.find("table", {"class": "wikitable"})
        #print(table)
#         header = table.find_all("th")
#         print(header)
 
        # Find all rows in the table
        rows = table.find_all('tr')
        #print(rows)
        # Extract column headers
        headers = [th.get_text().strip() for th in rows[0].find_all('th')]
        #print(headers)
        # Create an empty dataframe with columns
        df_insurance = pd.DataFrame(columns=headers)
        for rows in rows[1:]:
            column_data = [td.get_text().strip() for td in rows.find_all('td')]
            df_insurance = df_insurance.append(pd.Series(column_data, index=headers), ignore_index=True)
        #display(df_insurance)
        end_time = datetime.datetime.now()
        elapsed_time = end_time - start_time
        print("Total Scrapping for Population data Time:", elapsed_time)

 
    else:
         print("Failed to retrieve the webpage. Status code:", r.status_code)
 
 
    return df_insurance
 
 
```
